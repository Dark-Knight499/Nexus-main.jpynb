# Import Libraries

```python
import pandas as pd
import numpy as np
import sklearn as sk
```

# Load Files

```python
from sklearn.model_selection import train_test_split
df_train = pd.read_csv("train.csv")
df_train.head()
```

# Split in Train and Test

```python
to_remove = ["ID","SEQUENCE","FUNCTION","SYNONYMS","UNIPROID","PUBCHCID","Accession Number","TargetID"]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X1 = df_train.drop(columns=to_remove).drop(columns=["Target_Status"])
Y1 =le.fit_transform( df_train["Target_Status"])
df_test = pd.read_csv("test.csv")
X_train,X_test,Y_train,Y_test = train_test_split(X1,Y1,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_train_n = pd.get_dummies(X_train)
X_train_n
```

```python
from sklearn.neural_network import MLPClassifier 
clf = MLPClassifier(hidden_layer_sizes=(64, 32), activation='relu', solver='adam', max_iter=501, random_state=42)
clf.fit(X_train_n, Y_train)
```

```python

X_test_n = pd.get_dummies(X_test)
print(X_test_n.shape,X_train_n.shape)
missing_cols = set(X_train_n.columns) - set(X_test_n.columns)
for col in missing_cols:
    X_test_n[col] = 0
X_test_n = X_test_n[X_train_n.columns]
```

```python
clf.fit(X_test_n, Y_test)
```

```python
print(X_test_n.shape,X_train_n.shape)
```

```python
from sklearn.metrics import accuracy_score
result = clf.predict(X_test_n)
accuracy_score(result,Y_test)
```

```python
df_test_n = pd.get_dummies(df_test.drop(columns=["ID"]))
missing_cols = set(X_train_n.columns) - set(df_test_n)
for col in missing_cols:
    df_test_n[col] = 0
df_test_n = df_test_n[X_train_n.columns]
```

```python
df_test_n.shape
```

```python
df_test_n['Target_Status'] = clf.predict(df_test_n)
```

```python

predictions = pd.DataFrame({
    "ID" :df_test["ID"],
    "Target_Status": le.inverse_transform(df_test_n['Target_Status'])
})
predictions.to_csv('last_best_not_the least.csv', index=False)
```
